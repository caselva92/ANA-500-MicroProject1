## Loading relevant Libraries

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/carlosselva/Desktop/ANA-500/adult.csv") # Loading dataset for use
```

```python
df.shape
```

```python
df.dtypes #Looking at the type of data
```

```python
df.head() # Printing first five rows of data
```

```python
df.replace("?", pd.NA, inplace=True) # Replacing all question mark records with N/A to get a count of how many missing variables exist
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace('.', '_')
```

```python
df.head()
```

```python
df.isnull().sum() # Returning sum of all missing values
```

```python
workclass_counts = df.value_counts('workclass')
print(workclass_counts.to_dict())
```

```python
df = df.fillna({'workclass': 'Unknown'}) # Replacing all missing values for feature with Unknown 
```

```python
df.head()
```

```python
occupation_counts = df.value_counts('occupation')
print(occupation_counts.to_dict())
```

```python
df = df.fillna({'occupation': 'Unknown'}) # Replacing all missing values for feature with Unknown
```

```python
df.head()
```

```python
native_country_counts = df.value_counts('native_country')
print(native_country_counts.to_dict())
```

```python
most_frequent_country = df['native_country'].mode()[0]
df = df.fillna({'native_country': most_frequent_country})
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.head()
```

```python
df['sex'] = df['sex'].replace({'Male': 0, 'Female':1}) # Encoding labels for male and female for use in stastical analysis
```

```python
df.head(10)
```

```python
correlation_matrix = df.corr()
```

```python
print(correlation_matrix)
```

Based on the correlation matrix, sex and hours per week have the highest (negative) correlation. Secondly, education number and hours per week have the second highest correlation. These are possible features that could be used as predictors for income.
